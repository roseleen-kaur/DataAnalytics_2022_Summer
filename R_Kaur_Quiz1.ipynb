The price of a diamond is a very important attribute. It is decided by factoring in the 4C's: carat, color, cut and clarity of the diamond. The rarity of a diamond also adds to the price exponentially. The GIA (Gemology Institution of America) decides the value of a diamond.

The price is quantitative, discrete data and a numerical variable. It increases in line with the worth of the product.

```
CARAT
weight of the diamond (0.2--5.01)

```
When you go diamond jewelry shopping, or even in commercials, you will often hear people mention the carat of a diamond as one of its selling points. The carat of a diamond is its weight. A single carat is 100 milligrams, but it is also divide into 100 sub-points which is why the carat of diamond is given in a decimal value. The carat of a diamond is one of the factors that help decide its value. The higher the carat, the higher the price. It also indicates diamond quality, they both go up proportionally. Along with the price, the higher the carat the rarer the diamond. 

The carat is quantitative continuous varible since it's measured with numbers (weight) and can be broken down. It is a numerical variable.

```
CUT
quality of the cut (Fair, Good, Very Good, Premium, Ideal)

```

The cut of the diamond is one of, if not the most, important deciding factors of the value of the diamond. But what exactly is the cut of a diamond?

"[Diamond](https://www.diamonds.pro/education/cuts/) Cut is how well a diamond is cut and polished, including how well-proportioned the stone is, its depth and symmetry." A well cut diamond shines bright, while a poorly cut diamond is very dull. Which is why we have these 5 levels of the cut; fair being the lowest rank and ideal being the best. 

Since the cut can ordered and ranked into categories, it is a categorical ordinal variable.

```
COLOR
diamond color, from D (best) to J (worst)
```
![picture](https://drive.google.com/uc?id=12dHQp6gp4puM3NwrfdjbqWQ0Y-dI03lo)

The color of the diamond is another part of the process of deciding a diamond's value. In this data we are only paying attention to D-J coloring. THe color refers to how clear of yellow a diamond is. Usually the higher the quality the clearer it is and the more yellow, the lower the quality. Even in the data set below you'll see that the diamonds that are I-J are lower cut and lesser in price.

Like the cut, color can be ordered and ranked into categories, so it is a categorical ordinal variable.

```
CLARITY
a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
```
![picture](https://drive.google.com/uc?id=11cXQOXf4w2FVWwE0i5HCa1qPQHQPoHQm)

The clarity of a diamond is determined by if it has nay blemishes or inclusion. "[Natural](https://4cs.gia.edu/en-us/diamond-clarity/) diamonds are the result of carbon exposed to tremendous heat and pressure deep in the earth. This process can result in a variety of internal characteristics called ‘inclusions’ and external characteristics called ‘blemishes.’" Which is why this grading system exists; to categorize the diamonds in to a certain clarity.

[Definitions:](https://4cs.gia.edu/en-us/diamond-clarity/)

Flawless (FL): No inclusions and no blemishes visible under 10x magnification

Internally Flawless (IF): No inclusions visible under 10x magnification

Very, Very Slightly Included (VVS1 and VVS2): Inclusions so slight they are difficult for a skilled grader to see under 10x magnification

Very Slightly Included (VS1 and VS2): Inclusions are observed with effort under 10x magnification, but can be characterized as minor

Slightly Included (SI1 and SI2): Inclusions are noticeable under 10x magnification

Included (I1, I2, and I3): Inclusions are obvious under 10x magnification which may affect transparency and brilliance.

Clarity can ordered and ranked into categories, making it a categorical ordinal variable.

The depth is numerical data.

The table of the diamond is the top part you when you look down at it directly. it is the most imporant factor of a diamond's appearance. The table is what let's light into the diamond and also disperses it, to show the beautiful shine that everyone loves. The bigger the table the more the shine. ####Question 1####
What proportion of diamonds are between .30 and 1.08 carats?

````python
#I created a list out of the column of carats
Carats = df['carat'].tolist() 
#created empty list for answer 
new_list = []
#go through carats and check to see what's in between 
for x in Carats:
#I didn't do <=/ >= because it said in between, so I 
#interpreted that as non inclusive
  if x < 1.08 and x > 0.30:  
    new_list.append(x)
#get the percentage
print((len(new_list)/len(Carats))*100,'%')
````

###Question 3###
 How many of the diamonds have a carat value that is less than the mean carat value?

````python
import numpy as np
#find the average of carats value
mean = np.average(Carats)
print(f'Mean: {mean}')
#did a list comprehension: traversing through the values of carat
#if it is less thanthe mean value put in new array
less_than_mean =[[df['carat'].values[x] for x in range(len(Carats))  if Carats[x] < mean ]]
#again prints shape because the question was how many diamonds
print(np.shape(less_than_mean))
````

###Question 4###
 How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows:

_Fair / Good / Very Good / Premium / Ideal_

````python
import numpy as np
#made a list of the indexes so I could traverse through them, got it from:
#https://www.statology.org/pandas-index-to-list/
index_list = list(df.index.values)
#list of cut column
cut = df['cut'].tolist() 
#empty list
dia_list = []
#traverse through the indexes
for x in range(len(index_list)):
  if cut[x] == 'Premium': #if premium append
    dia_list.append(x)
  if cut[x] == 'Ideal': #if ideal append
    dia_list.append(x)
#print shape because we want how many diamonds
print(f'Number of Diamonds: {np.shape(dia_list)}')
````
